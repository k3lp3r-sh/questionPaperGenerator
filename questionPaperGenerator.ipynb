```python
import time
```

```python
langfuse = Langfuse(
    public_key=os.getenv("LANGFUSE_PUBLIC_KEY"),
    secret_key=os.getenv("LANGFUSE_SECRET_KEY"),
    host=os.getenv("LANGFUSE_HOST")
)

trace = langfuse.trace(name="generate_questions", user_id="user-1")

# Start timing
start_time = time.time()

# Your prompt
prompt = f"""Consider this to be your dataset: {question_bank} 
it's a dump of questions in the topic {chapter_dict[9]}. 
You are a teacher, testing students with a test. 
Generate 15 questions similar to or from the dataset I've provided. 
You are allowed to create questions new questions by combining two or more questions. 
You are allowed to generate critical reasoning questions. 
Each question must be a multiple choice question with four or more options. 
Include Statement Type questions, Assertion Reasoning type questions Match the following type questions etc."""

# Add prompt as a generation span
span = trace.span(name="generate_question_paper", input=prompt)

# Call Gemini API
response = client.models.generate_content(model="gemini-2.0-flash", contents=prompt)

# Stop timer
latency = time.time() - start_time

# Set output and duration
span.output = response.text
span.end_time = start_time + latency
span.level = "INFO"

langfuse.flush()

# Capture the question paper
qp = response.text

```
